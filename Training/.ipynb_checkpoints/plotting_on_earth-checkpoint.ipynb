```python
import matplotlib.pyplot as plt
import geopandas as gpd
import psycopg2
from jproperties import Properties
import os
```

```python
root_path = '/Users/sidpai/Desktop/Sem_2'
os.chdir(root_path)

properties_filename = 'EUDR_MTP/Training/credentials.properties'
configs = Properties()
with open(properties_filename, 'rb') as config_file:
    configs.load(config_file)
```

```python
world = gpd.read_file(gpd.datasets.get_path("naturalearth_lowres"))
```

```python
host = configs.get("host").data
database = configs.get("database").data
user = configs.get("user").data
password = configs.get("password").data
port = configs.get("port").data
```

```python
connection = psycopg2.connect(host=host, database=database, user=user, password=password, port=port)
cursor = connection.cursor()
query = """SELECT centroid_longitude,centroid_latitude,radius from eudr.forest_coordinates where year = 2016"""
cursor.execute(query)
rows_2016 = cursor.fetchall()
connection.commit()
cursor.close()
connection.close()
```

```python
connection = psycopg2.connect(host=host, database=database, user=user, password=password, port=port)
cursor = connection.cursor()
query = """SELECT centroid_longitude,centroid_latitude,radius from eudr.forest_coordinates where year = 2020"""
cursor.execute(query)
rows_2020 = cursor.fetchall()
connection.commit()
cursor.close()
connection.close()
```

```python
figure, axes = plt.subplots()
plt.xlim(-180,180)
plt.ylim(-90,90)
```

```python
world.plot(ax=axes,color='white',edgecolor='grey')
```

```python
for i in rows_2016:
    cent = (i[0],i[1])
    rad = i[2]
    c1 = plt.Circle(cent, rad, color='r')
    axes.add_patch(c1)
    
for i in rows_2020:
    cent = (i[0],i[1])
    rad = i[2]
    c1 = plt.Circle(cent, rad, color='g')
    axes.add_patch(c1)
```

```python
axes.set_aspect('equal')
```
